# **Deep Causality Graph Model**
# TE Process DCG Model with Group Lasso

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.utils.data import Dataset, DataLoader
import os
import pandas as pd

# 数据集类
class TEDataset(Dataset):
    def __init__(self, data_folder, window_size=10):
        self.data = self._load_all_data(data_folder)
        self.window_size = window_size

    def _load_all_data(self, data_folder):
        all_data = []
        for filename in os.listdir(data_folder):
            if filename.endswith('.dat'):
                file_path = os.path.join(data_folder, filename)
                try:
                    data = pd.read_csv(file_path, sep='\s+', header=None)
                    if data.shape[1] > data.shape[0]:
                        data = data.T
                    data = data.iloc[:, :31]
                    print(f"Loaded {filename} with shape: {data.shape}")
                    all_data.append(data)
                except Exception as e:
                    print(f"Error loading {filename}: {e}")

        if not all_data:
            raise ValueError("No valid data files found in the folder")

        combined_data = pd.concat(all_data, axis=0, ignore_index=True)
        combined_data = (combined_data - combined_data.mean()) / combined_data.std()
        print(f"Combined data shape: {combined_data.shape}")
        return combined_data

    def __len__(self):
        return len(self.data) - self.window_size

    def __getitem__(self, idx):
        window = self.data.iloc[idx:idx+self.window_size].values
        x = window[:-1]  # 前T-1个时间步
        y = window[-1]   # 最后一个时间步
        return torch.FloatTensor(x), torch.FloatTensor(y)

# GRU特征提取器
class GRUFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)

    def forward(self, x):
        output, hidden = self.gru(x)
        return output[:, -1, :]

# MLP特征提取器
class MLPFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.Tanh()
        )
        self.input_size = input_size  # 记录输入维度

    def forward(self, x):
        return self.mlp(x)

# 预测输出单元
class PredictionUnit(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.mean_net = nn.Sequential(
            nn.Linear(input_size, input_size//2),
            nn.Tanh(),
            nn.Linear(input_size//2, 1)
        )
        self.var_net = nn.Sequential(
            nn.Linear(input_size, input_size//2),
            nn.Tanh(),
            nn.Linear(input_size//2, 1),
            nn.Softplus()
        )

    def forward(self, x):
        mean = self.mean_net(x)
        var = self.var_net(x)
        return mean, torch.log(var + 1e-6)

# 深度因果图模型
class DeepCausalityGraph(nn.Module):
    def __init__(self, num_vars, hidden_size=30):
        super().__init__()
        self.num_vars = num_vars
        self.hidden_size = hidden_size

        self.gru_extractors = nn.ModuleList([
            GRUFeatureExtractor(num_vars, hidden_size)
            for _ in range(num_vars)
        ])

        self.mlp_extractors = nn.ModuleList([
            MLPFeatureExtractor(num_vars-1, hidden_size)
            for _ in range(num_vars)
        ])

        self.predictors = nn.ModuleList([
            PredictionUnit(hidden_size * 2)
            for _ in range(num_vars)
        ])

    def forward(self, x):
        batch_size = x.shape[0]
        means = []
        log_vars = []

        for i in range(self.num_vars):
            # GRU特征提取
            gru_feat = self.gru_extractors[i](x)

            # MLP特征提取
            current_step = x[:, -1, :]
            other_vars = torch.cat([current_step[:, :i], current_step[:, i+1:]], dim=1)
            mlp_feat = self.mlp_extractors[i](other_vars)

            # 合并特征
            combined_feat = torch.cat([gru_feat, mlp_feat], dim=1)

            # 预测
            mean, log_var = self.predictors[i](combined_feat)
            means.append(mean)
            log_vars.append(log_var)

        return torch.cat(means, dim=1), torch.cat(log_vars, dim=1)

# 修改后的Group Lasso正则化
def group_lasso_penalty(model):
    penalty = 0
    for i, mlp in enumerate(model.mlp_extractors):
        weight_matrix = mlp.mlp[0].weight
        num_features = mlp.input_size

        for j in range(num_features):
            if j >= i:
                j_actual = j + 1  # 跳过自身的变量
            else:
                j_actual = j
            w = weight_matrix[:, j]
            penalty += torch.sqrt(torch.sum(w**2))

    return penalty

# 训练函数
def train_dcg(model, train_loader, optimizer, lambda_reg=0.5, epochs=100, device='cuda' if torch.cuda.is_available() else 'cpu'):
    model = model.to(device)
    model.train()

    for epoch in range(epochs):
        total_loss = 0
        for batch_x, batch_y in train_loader:
            batch_x = batch_x.to(device)
            batch_y = batch_y.to(device)

            optimizer.zero_grad()

            # 前向传播
            mean_pred, log_var_pred = model(batch_x)

            # 计算损失
            reconstruction_loss = 0.5 * (log_var_pred +
                                       (batch_y - mean_pred)**2 / torch.exp(log_var_pred))
            reconstruction_loss = reconstruction_loss.mean()

            # Group Lasso正则化
            reg_loss = lambda_reg * group_lasso_penalty(model)

            # 总损失
            loss = reconstruction_loss + reg_loss

            # 反向传播
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}')

            # 保存检查点
            checkpoint = {
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': total_loss/len(train_loader)
            }
            torch.save(checkpoint, f'dcg_model_checkpoint_epoch_{epoch+1}.pth')

def main():
    data_folder = '/content/drive/My Drive/DCG_data'
    window_size = 10
    hidden_size = 30
    num_vars = 31
    batch_size = 32
    learning_rate = 0.001
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    try:
        print(f"Using device: {device}")

        print("Creating dataset...")
        dataset = TEDataset(data_folder, window_size)
        print("Creating data loader...")
        train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

        print("Testing data shapes...")
        x, y = next(iter(train_loader))
        print(f"Input batch shape: {x.shape}")
        print(f"Target batch shape: {y.shape}")

        print("Creating model...")
        model = DeepCausalityGraph(num_vars, hidden_size)

        optimizer = optim.Adam(model.parameters(), lr=learning_rate)

        print("Starting training...")
        train_dcg(model, train_loader, optimizer, device=device)

        print("Saving final model...")
        torch.save({
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'final_loss': None
        }, 'dcg_model_final.pth')
        print("Training completed!")

    except Exception as e:
        print(f"Error in main function: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    print("Checking data folder contents:")
    data_folder = '/content/drive/My Drive/DCG_data'
    try:
        from google.colab import drive
        drive.mount('/content/drive')
        files = os.listdir(data_folder)
        print(f"Files found: {files}")
    except Exception as e:
        print(f"Error accessing data folder: {e}")

    main()
```

# **Causal Network Graph**

```python
import torch
import torch.nn as nn
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 保持模型结构定义与训练时完全一致
class GRUFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)

    def forward(self, x):
        output, hidden = self.gru(x)
        return output[:, -1, :]

class MLPFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.Tanh()
        )
        self.input_size = input_size

    def forward(self, x):
        return self.mlp(x)

class PredictionUnit(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.mean_net = nn.Sequential(
            nn.Linear(input_size, input_size//2),
            nn.Tanh(),
            nn.Linear(input_size//2, 1)
        )
        self.var_net = nn.Sequential(
            nn.Linear(input_size, input_size//2),
            nn.Tanh(),
            nn.Linear(input_size//2, 1),
            nn.Softplus()
        )

    def forward(self, x):
        mean = self.mean_net(x)
        var = self.var_net(x)
        return mean, torch.log(var + 1e-6)

class DeepCausalityGraph(nn.Module):
    def __init__(self, num_vars, hidden_size=30):
        super().__init__()
        self.num_vars = num_vars
        self.hidden_size = hidden_size

        self.gru_extractors = nn.ModuleList([
            GRUFeatureExtractor(num_vars, hidden_size)
            for _ in range(num_vars)
        ])

        self.mlp_extractors = nn.ModuleList([
            MLPFeatureExtractor(num_vars-1, hidden_size)
            for _ in range(num_vars)
        ])

        self.predictors = nn.ModuleList([
            PredictionUnit(hidden_size * 2)
            for _ in range(num_vars)
        ])

def extract_causal_strengths(model):
    """提取因果强度并进行归一化"""
    causal_strengths = np.zeros((model.num_vars, model.num_vars))

    for i, mlp in enumerate(model.mlp_extractors):
        weight_matrix = mlp.mlp[0].weight.detach().cpu().numpy()
        influence_strength = np.sqrt(np.sum(weight_matrix**2, axis=0))

        # 将影响强度填入邻接矩阵
        causal_strengths[i, :i] = influence_strength[:i]
        causal_strengths[i, i+1:] = influence_strength[i:]

    # 归一化强度
    max_strength = np.max(causal_strengths)
    if max_strength > 0:
        causal_strengths = causal_strengths / max_strength

    return causal_strengths

def visualize_causal_graph(model, threshold=0.1):
    """可视化因果图"""
    # 提取因果强度
    causal_strengths = extract_causal_strengths(model)

    # 创建有向图
    G = nx.DiGraph()

    # 添加节点
    for i in range(model.num_vars):
        G.add_node(f"Var {i+1}")

    # 记录添加的边数
    edge_count = 0

    # 添加边
    for i in range(model.num_vars):
        for j in range(model.num_vars):
            if i != j and causal_strengths[i, j] > threshold:
                G.add_edge(f"Var {j+1}", f"Var {i+1}",
                          weight=causal_strengths[i, j])
                edge_count += 1

    print(f"添加的边数量: {edge_count}")

    # 设置绘图参数
    plt.figure(figsize=(15, 15))
    pos = nx.spring_layout(G, k=2.0, iterations=50)

    # 绘制节点
    nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                          node_size=1000, alpha=0.7)
    nx.draw_networkx_labels(G, pos)

    # 绘制边
    if edge_count > 0:
        edges = G.edges()
        weights = [G[u][v]['weight'] for u, v in edges]
        nx.draw_networkx_edges(G, pos, edgelist=edges,
                             width=[w*3 for w in weights],
                             alpha=0.6,
                             edge_color='red',
                             arrows=True,
                             arrowsize=20)

    plt.title("Causal Network Graph")
    plt.axis('off')
    plt.tight_layout()

    # 打印一些统计信息
    print("\n因果关系强度统计:")
    print(f"最大强度: {np.max(causal_strengths):.4f}")
    print(f"平均强度: {np.mean(causal_strengths):.4f}")
    print(f"最小强度: {np.min(causal_strengths):.4f}")

    # 显示最强的几个因果关系
    print("\n最强的5个因果关系:")
    flat_idx = np.argsort(causal_strengths.flatten())[-6:]  # 多取一个以防自身关系
    for idx in flat_idx:
        i, j = idx // causal_strengths.shape[1], idx % causal_strengths.shape[1]
        if i != j:  # 跳过自身关系
            print(f"Var {j+1} -> Var {i+1}: {causal_strengths[i,j]:.4f}")

    plt.show()
    return plt

def visualize_from_checkpoint(model_path, threshold=0.5):
    """从检查点加载模型并可视化"""
    try:
        # 创建模型
        model = DeepCausalityGraph(31, 30)

        # 加载权重
        checkpoint = torch.load(model_path, map_location='cpu')
        model.load_state_dict(checkpoint['model_state_dict'])
        model.eval()

        # 可视化
        return visualize_causal_graph(model, threshold)

    except Exception as e:
        print(f"Error: {e}")
        return None

# 使用示例
if __name__ == "__main__":
    model_path = 'dcg_model_final.pth'  # 确保这是您的模型文件路径
    plt = visualize_from_checkpoint(model_path, threshold=0.5)
```

# **TE Process Causality Matrix Analysis**

```python
def extract_causal_matrix(model, threshold=None):
    """提取因果矩阵"""
    num_vars = model.num_vars
    causal_matrix = np.zeros((num_vars, num_vars))

    # 从模型中提取权重
    for i, mlp in enumerate(model.mlp_extractors):
        weights = mlp.mlp[0].weight.data.cpu().numpy()
        for j in range(weights.shape[1]):
            if j >= i:
                j_actual = j + 1
            else:
                j_actual = j
            # 计算因果强度
            causal_strength = np.linalg.norm(weights[:, j])
            causal_matrix[i, j_actual] = causal_strength

    # 如果提供了阈值，进行二值化
    if threshold is not None:
        if threshold == 'auto':
            # 自动确定阈值（例如使用平均值加标准差）
            threshold = np.mean(causal_matrix) + np.std(causal_matrix)
        causal_matrix = (causal_matrix > threshold).astype(int)

    return causal_matrix

def print_causality_matrix(model, threshold='auto', save_path=None):
    """打印并可选保存因果矩阵"""
    # 提取因果矩阵
    causal_matrix = extract_causal_matrix(model, threshold=threshold)

    # 创建DataFrame
    df = pd.DataFrame(
        causal_matrix,
        index=range(1, model.num_vars + 1),
        columns=range(1, model.num_vars + 1)
    )

    # 打印矩阵
    print("\nThe causality matrix of TE process:")
    print(df.to_string(index=True, header=True))

    # 如果指定了保存路径，保存到文件
    if save_path:
        df.to_csv(save_path)
        print(f"\nCausality matrix saved to: {save_path}")

    return df

# 在主函数或评估代码中使用
def evaluate_causality():
    # 设置参数
    num_vars = 31
    hidden_size = 30
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    try:
        # 加载模型
        print("Loading model...")
        model = DeepCausalityGraph(num_vars, hidden_size)
        checkpoint = torch.load('dcg_model_final.pth', map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.to(device)

        # 打印因果矩阵
        causality_df = print_causality_matrix(
            model,
            threshold='auto',
            save_path='results/causality_matrix.csv'
        )

        # 分析因果关系
        print("\nAnalyzing causal relationships:")
        non_zero = np.where(causality_df.values != 0)
        for i, j in zip(*non_zero):
            print(f"Variable {i+1} -> Variable {j+1}")

    except Exception as e:
        print(f"Error in evaluation: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    evaluate_causality()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import DataLoader

def evaluate_model(model, test_loader, device):
    """评估模型性能"""
    model.eval()
    total_loss = 0
    predictions = []
    actuals = []

    with torch.no_grad():
        for batch_x, batch_y in test_loader:
            batch_x = batch_x.to(device)
            batch_y = batch_y.to(device)

            # 预测
            mean_pred, log_var_pred = model(batch_x)

            # 计算损失
            loss = 0.5 * (log_var_pred + (batch_y - mean_pred)**2 / torch.exp(log_var_pred))
            loss = loss.mean()

            total_loss += loss.item()

            predictions.append(mean_pred.cpu().numpy())
            actuals.append(batch_y.cpu().numpy())

    avg_loss = total_loss / len(test_loader)
    predictions = np.concatenate(predictions)
    actuals = np.concatenate(actuals)

    # 计算R²分数
    r2_scores = []
    for i in range(predictions.shape[1]):
        r2 = 1 - np.sum((actuals[:,i] - predictions[:,i])**2) / np.sum((actuals[:,i] - actuals[:,i].mean())**2)
        r2_scores.append(r2)

    return avg_loss, r2_scores, predictions, actuals

def extract_causal_matrix(model):
    """提取因果矩阵"""
    causal_matrix = np.zeros((model.num_vars, model.num_vars))

    for i, mlp in enumerate(model.mlp_extractors):
        weights = mlp.mlp[0].weight.data.cpu().numpy()
        for j in range(weights.shape[1]):
            if j >= i:
                j_actual = j + 1
            else:
                j_actual = j
            causal_strength = np.linalg.norm(weights[:, j])
            causal_matrix[i, j_actual] = causal_strength

    return causal_matrix

def plot_causal_matrix(causal_matrix, save_path='results/causal_matrix.png'):
    """绘制并保存因果矩阵图"""
    plt.figure(figsize=(15, 12))
    sns.heatmap(causal_matrix,
                cmap='RdBu_r',
                center=0)
    plt.title('Causal Relationship Matrix')
    plt.xlabel('Cause Variables')
    plt.ylabel('Effect Variables')
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

def main_evaluation():
    # 设置参数
    data_folder = '/content/drive/My Drive/DCG_data'
    window_size = 10
    hidden_size = 30
    num_vars = 31
    batch_size = 32
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    try:
        # 创建数据集和加载器
        print("Loading test dataset...")
        test_dataset = TEDataset(data_folder, window_size)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        # 加载模型
        print("Loading model...")
        model = DeepCausalityGraph(num_vars, hidden_size)
        checkpoint = torch.load('dcg_model_final.pth', map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.to(device)

        # 评估模型
        print("\nEvaluating model performance...")
        avg_loss, r2_scores, predictions, actuals = evaluate_model(model, test_loader, device)
        print(f"Average Loss: {avg_loss:.4f}")
        print("\nR² scores for each variable:")
        for i, r2 in enumerate(r2_scores):
            print(f"Variable {i}: {r2:.4f}")

        # 提取并绘制因果矩阵
        print("\nExtracting causal matrix...")
        causal_matrix = extract_causal_matrix(model)

        # 创建结果目录
        os.makedirs('results', exist_ok=True)

        print("Plotting causal matrix...")
        plot_causal_matrix(causal_matrix)

        # 保存评估结果
        print("\nSaving evaluation results...")
        np.savez('results/evaluation_results.npz',
                causal_matrix=causal_matrix,
                r2_scores=r2_scores,
                avg_loss=avg_loss,
                predictions=predictions,
                actuals=actuals)

        # 打印最强因果关系
        print("\nTop 10 strongest causal relationships:")
        flat_indices = np.argsort(causal_matrix.flatten())[-10:]
        for idx in flat_indices:
            i, j = np.unravel_index(idx, causal_matrix.shape)
            print(f"Variable {i} -> Variable {j}: Strength = {causal_matrix[i,j]:.4f}")

    except Exception as e:
        print(f"Error in evaluation: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main_evaluation()
```

# **VCI**

```python
def calculate_vci(model, test_loader, device, window_size=50, alpha_c=0.95):
    """
    计算变量贡献度指标(VCI)

    参数:
    - model: 训练好的深度因果图模型
    - test_loader: 测试数据加载器
    - window_size: 计算VCI的时间窗口大小
    - alpha_c: 概率检测指标的阈值
    """
    model.eval()
    num_vars = model.num_vars

    # 存储每个变量的异常计数
    anomaly_counts = np.zeros(num_vars)
    total_samples = 0

    with torch.no_grad():
        for batch_x, batch_y in test_loader:
            batch_x = batch_x.to(device)
            batch_y = batch_y.to(device)

            # 获取模型预测
            mean_pred, log_var_pred = model(batch_x)

            # 计算每个变量的概率检测指标
            for i in range(num_vars):
                mse = (batch_y[:, i] - mean_pred[:, i])**2
                var = torch.exp(log_var_pred[:, i])
                prob_anomaly = 1 - torch.exp(-0.5 * mse / var)

                # 统计超过阈值的样本数
                anomaly_counts[i] += torch.sum(prob_anomaly > (1 - alpha_c)).item()

            total_samples += batch_y.size(0)

    # 计算VCI
    vci_values = anomaly_counts / np.sum(anomaly_counts)
    return vci_values

def plot_vci_bar(vci_values, threshold=0.05, save_path=None):
    """
    绘制VCI柱状图

    参数:
    - vci_values: VCI值数组
    - threshold: 阈值线
    - save_path: 保存图片的路径（可选）
    """
    n_variables = len(vci_values)
    variables = range(1, n_variables + 1)

    plt.figure(figsize=(10, 6))

    # 绘制柱状图
    plt.bar(variables, vci_values, color='#0A2D8F', width=0.6)
    plt.axhline(y=threshold, color='red', linestyle='--', linewidth=1, label='Threshold')

    # 设置标题和标签
    plt.xlabel("变量", fontsize=12)
    plt.ylabel("VCI", fontsize=12)

    # 设置坐标轴范围和刻度
    plt.ylim(0, max(vci_values) * 1.1)
    plt.xticks(np.arange(0, n_variables+1, 5))
    plt.yticks(np.arange(0, max(vci_values)*1.1, 0.05))

    # 去除顶部和右侧边框
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # 添加网格线
    plt.grid(True, axis='y', linestyle='--', alpha=0.3)

    plt.tight_layout()

    # 保存图片
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

    plt.show()

    # 打印超过阈值的变量
    fault_vars = np.where(vci_values > threshold)[0] + 1
    print("\nFault-related variables:")
    print(f"Variables {fault_vars} (VCI > {threshold})")

def evaluate_vci():
    """评估VCI和绘图的主函数"""
    # 设置参数
    num_vars = 31
    hidden_size = 30
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    try:
        # 加载模型和数据
        print("Loading model and data...")
        model = DeepCausalityGraph(num_vars, hidden_size)
        checkpoint = torch.load('dcg_model_final.pth', map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.to(device)

        # 创建测试数据加载器
        test_dataset = TEDataset('/content/drive/My Drive/DCG_data', window_size=10)
        test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

        # 计算VCI
        print("Calculating VCI values...")
        vci_values = calculate_vci(model, test_loader, device)

        # 绘制VCI图
        print("Plotting VCI values...")
        plot_vci_bar(
            vci_values,
            threshold=1/num_vars,  # 使用1/n作为阈值
            save_path='results/vci_plot.png'
        )

    except Exception as e:
        print(f"Error in VCI evaluation: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    evaluate_vci()
```
